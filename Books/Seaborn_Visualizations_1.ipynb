# Heat maps

```python
df_heatmap = sns.load_dataset('mpg')
df_heatmap.head()
```

```python
df_heatmap = df_heatmap.drop('origin', axis=1)
```

```python
df_heatmap = df_heatmap.drop('name', axis=1)
```

```python
df_heatmap.corr()
```

```python
my_corr_mat = df_heatmap.corr()
```

```python
sns.heatmap(my_corr_mat);
```

```python
sns.heatmap(my_corr_mat, cmap='plasma');
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, cmap='plasma');
```

```python
plt.figure(figsize=(12,6), dpi=250)
sns.heatmap(my_corr_mat, cmap='plasma', annot=True, fmt='.3f');
```

```python
plt.figure(figsize=(12,6), dpi=250)
sns.heatmap(my_corr_mat, annot=True, fmt='.3f', linewidth=2, linecolor='red', cmap='magma');
```

```python
plt.figure(figsize=(12,6), dpi=250)
sns.heatmap(my_corr_mat, cmap='plasma', annot=True, fmt='.3f', cbar=False);
```

```python
plt.figure(figsize=(8,8), dpi=250)
sns.heatmap(my_corr_mat, cmap='plasma', annot=True, fmt='.3f', square=True);
```

```python
plt.figure(figsize=(12,6), dpi=250)
sns.heatmap(my_corr_mat, cmap='plasma', annot=True, mask=np.triu(my_corr_mat));
```

# Boxen or Letter Value Plots

```python
df_mush = pd.read_csv('secondary_data.csv', sep=';')
df_mush.head()
```

```python
sns.boxplot(x=df_mush['cap-diameter']); 
```

```python
sns.boxenplot(x=df_mush['cap-diameter']); 
```

```python
plt.figure(figsize=(12,6), dpi=250)
sns.boxenplot(x=df_mush['cap-diameter']); 
```

```python
fig, ax = plt.subplots(figsize=(12,6), dpi=250)
fig = sns.boxplot(x=df_mush['cap-diameter'])
plt.show()
```

## Mushroom Cap Diameter By Habitat

```python
fig, ax = plt.subplots(figsize=(10,5), dpi=250)
fig = sns.boxenplot(x='habitat', y='cap-diameter', data=df_mush)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5), dpi=250)
fig = sns.boxenplot(x='habitat', y='cap-diameter', data=df_mush, hue='has-ring')
plt.show()
```

```python
fig1, ax = plt.subplots(figsize=(10,5), dpi=250)
fig1 = sns.boxenplot(x='habitat', y='cap-diameter', data=df_mush, hue='has-ring', order=sorted(df_mush['habitat'].unique(),reverse=True))
plt.show()
```

```python
fig1, ax = plt.subplots(figsize=(10,5), dpi=250)
fig1 = sns.boxenplot(x='habitat', y='cap-diameter', data=df_mush, hue='has-ring', 
                     order=sorted(df_mush['habitat'].unique(),reverse=True),  palette='coolwarm')
plt.show()
```

## Facet Grid Analysis

```python
df_dia1 = sns.load_dataset('diamonds')
df_dia1.head()
```

```python
plt.figure(figsize=(12,6), dpi=250)
g = sns.FacetGrid(df_dia1, col='cut', hue='clarity', palette='coolwarm', col_wrap=3, legend_out=False)
g.map(plt.scatter, "carat", "price", s=70)
g.add_legend() ;
```
