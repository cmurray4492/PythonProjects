# Pandas Visualizations Customizations

```python
import pandas as pd
```

```python
df2 = pd.read_csv('df2.csv')
```

```python
df2.head()
```

```python
df2.plot.line();
```

```python
df2['c'].plot.line(figsize=(10,3),ls='--', c='red', lw=4);
```

```python
title = "MY PLOT TITLE"
xlabel = "My X Data"
ylabel = "My Y Data"
```

```python
ax = df2['c'].plot.line(figsize=(10,3),ls='--', c='red', lw=4, title=title)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```python
ax = df2.plot();
ax.legend(loc=0, bbox_to_anchor=(1.0, 1.0))
```
