# Pandas Visualizations

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('df1.csv', index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2.csv')
```

```python
df2.head()
```

## Histograms

```python
df1['A'].plot.hist();
```

```python
df1['A'].plot.hist(bins=15,edgecolor='k').autoscale(enable=True, axis='both', tight=True)
```

```python
df1['A'].plot.hist(grid=True,bins=15,edgecolor='k').autoscale(enable=True, axis='both', tight=True)
```

## Barplots

```python
df2.plot.bar();
```

```python
df2.plot.barh(stacked=True);
```

## LIne Plots

```python
df2.plot.line(y=['a', 'b', 'c'], figsize=(10,4), lw=4);
```

## Area Plots

```python
df2.plot.area(alpha=0.5); # stacked=False - mixes colors 
```

## Scatter Plots

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm'); # c = color by
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*15, c='D', cmap='coolwarm', alpha=0.75); # is size of marker
```

## Box Plot

```python
df2.plot.box();
```

```python
df2.boxplot();
```

## KDE Plot

```python
df2['a'].plot.kde();
```

```python
df2.plot.kde();
```

## Hex Bin Plot

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a', 'b'])
```

```python
df.plot.scatter(x='a', y='b');
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25, cmap='Oranges');
```
