# Simple Linear Regression with Stats Models

```python
import pandas as pd
import numpy as np 
import statsmodels.api as sm
```

```python
diamonds = pd.read_csv('Data/Diamonds Prices2022.csv')
```

```python
diamonds.head()
```

```python
diamonds.shape
```

```python
diamonds.info()
```

```python
diamonds.describe().T
```

```python
X = sm.add_constant(diamonds['carat'])
y = diamonds['price']
```

```python
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```
